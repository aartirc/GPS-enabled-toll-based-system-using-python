```python
import folium
import random
from geopy.distance import geodesic
from datetime import datetime, timedelta

# Define routes
routes = {
    "Mumbai to Pune": [(19.0760, 72.8777), (18.5204, 73.8567)],
    "Bhopal to Indore": [(23.2599, 77.4126), (22.7196, 75.8577)],
    "Wardha to Bhandara": [(20.7453, 78.6022), (21.1667, 79.6558)]
}

# Function to generate random toll locations along a route
def generate_toll_locations(start, end, num_tolls):
    tolls = []
    for _ in range(num_tolls):
        random_factor = random.uniform(0, 1)
        lat = start[0] + random_factor * (end[0] - start[0])
        lon = start[1] + random_factor * (end[1] - start[1])
        tolls.append((lat, lon))
    return tolls

# Create a folium map centered around India
map_center = (21.1466, 79.0888)  # Near the geographical center of India
map_ = folium.Map(location=map_center, zoom_start=6)

# Function to calculate distance and speed
def calculate_distance_speed(start, end, speed_kmh):
    distance_km = geodesic(start, end).kilometers
    time_hours = distance_km / speed_kmh
    return distance_km, time_hours

# Function to simulate a journey
def simulate_journey(route_name, start, end, speed_kmh, tolls):
    total_distance = 0
    total_time = 0
    toll_charges = []
    current_time = datetime.now()

    # Add start point
    folium.Marker(location=start, popup=f"{route_name} Start\nTime: {current_time.strftime('%H:%M:%S')}", icon=folium.Icon(color='green')).add_to(map_)

    for i, toll in enumerate(tolls):
        distance, time_hours = calculate_distance_speed(start, toll, speed_kmh)
        charge = random.randint(50, 150)  # Random toll charge between 50 and 150

        total_distance += distance
        total_time += time_hours
        current_time += timedelta(hours=time_hours)

        toll_charges.append((toll, charge))

        # Add toll marker
        folium.Marker(location=toll, popup=f"Toll {i+1}\nCharge: ₹{charge}\nTime: {current_time.strftime('%H:%M:%S')}", icon=folium.Icon(color='blue')).add_to(map_)

        start = toll

    # Add end point
    distance, time_hours = calculate_distance_speed(start, end, speed_kmh)
    total_distance += distance
    total_time += time_hours
    current_time += timedelta(hours=time_hours)

    folium.Marker(location=end, popup=f"{route_name} End\nTime: {current_time.strftime('%H:%M:%S')}", icon=folium.Icon(color='red')).add_to(map_)

    folium.PolyLine([start] + tolls + [end], color="blue", weight=2.5, opacity=1).add_to(map_)

    return total_distance, total_time, toll_charges

# Add routes and simulate journeys
for route, (start, end) in routes.items():
    # Generate toll locations
    tolls = generate_toll_locations(start, end, num_tolls=5)

    # Define vehicle speed (average speed in km/h)
    vehicle_speed = 60  # km/h

    # Simulate the journey
    total_distance, total_time, toll_charges = simulate_journey(route, start, end, vehicle_speed, tolls)

    print(f"\nRoute: {route}")
    print(f"Total Distance: {total_distance:.2f} km")
    print(f"Total Time: {total_time:.2f} hours")
    print("Toll Charges:")
    for i, (toll, charge) in enumerate(toll_charges):
        print(f"  Toll {i+1}: Charge: ₹{charge}, Location: {toll}")

# Save the map to an HTML file and display it
map_.save("toll_routes.html")
map_
```
